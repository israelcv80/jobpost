```python
import os
#import md5
import hashlib
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
files = os.listdir("resume-job-posting-nlp-project\data\html_job_postings\html_job_postings")
```

```python
#create empty dataframe
webset = {"init":"init"}
datafdata = None
```

```python
for file in files:
    print("resume-job-posting-nlp-project/data/html_job_postings/html_job_postings/"+file)
    tempstring = open("resume-job-posting-nlp-project/data/html_job_postings/html_job_postings/"+file, 'r',encoding='UTF-8')
    lineslines =tempstring.readlines() 
    joinedlines = "".join(lineslines)
#     print(joinedlines)
    h = hashlib.md5()
    h.update(joinedlines.encode())
    valuehash = h.hexdigest()
    print(valuehash in webset)
    if valuehash in webset :
        print("already")
    else:
        print(h.hexdigest())
        print(joinedlines)
        print("--------------------------------------------------------")
        webset[h.hexdigest()]=joinedlines
        #print("adding")
        
        
#     print(h.hexdigest())
#     print(tempstring.read())
    
```
